```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (20.0,10.0)


#reading data
data= pd.read_csv('C:\\Users\\pooja\\Desktop\\headbrain.csv')
print(data.shape)
data.head()
```

```python
#collecting X and Y 
x = data['Head Size(cm^3)'].values
y = data['Brain Weight(grams)'].values
```

```python
#mean X and Y
mean_x = np.mean(x)
mean_y = np.mean(y)

#total no. of values
n = len(x)

# using the formula to calculate b1 and b0
numer = 0
denom = 0
for i in range(n):
    numer += (x[i] -np.mean(x)) * (y[i] - np.mean(y))
    denom += (x[i] - np.mean(x)) ** 2 

b1 = numer / denom
b0 = mean_y - (b1 *np.mean(x))


# print coefficient 
#here b1 = slope and b0 = intercept
print(b1,b0)
```

```python
# plotting values and regression line

max_x = np.mean(x) + 100
min_x = np.mean(y) - 100


# calculating line values x and y 
x = np.linspace(min_x, max_x, 1000)
y = b0 + b1 * x

# plotting scatter
plt.scatter(x,y, c = '#ef5423', label = 'scatter plot')

# ploting line
plt.plot(x,y, color = '#58b970', label = 'regression line')

#set the y axis range
plt.ylim(0,1600) #adjust
plt.xlim(0,4500)

plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend(loc='upper left')  # adjust the legend position
plt.show()
```
